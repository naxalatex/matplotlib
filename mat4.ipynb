```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample Dataset: Yearly Revenue and Ad Spend
df_company = pd.DataFrame({
    'Year': [2018, 2019, 2020, 2021, 2022, 2023],
    'Revenue': [50000, 60000, 70000, 85000, 95000, 110000],
    'Ad Spend': [5000, 7000, 9000, 10000, 11000, 13000]
})

# Create a figure with 2 rows and 1 column of subplots
plt.figure(figsize=(8, 8))

# Top plot: Line chart (Year vs Revenue)
plt.subplot(2, 1, 1)
plt.plot(df_company['Year'], df_company['Revenue'], marker='o', color='green', linestyle='-')
plt.title("Yearly Revenue Growth")
plt.xlabel("Year")
plt.ylabel("Revenue (in $)")
plt.grid(True)

# Bottom plot: Scatter chart (Ad Spend vs Revenue)
plt.subplot(2, 1, 2)
plt.scatter(df_company['Ad Spend'], df_company['Revenue'], color='blue', edgecolor='black')
plt.title("Advertising Spend vs Revenue")
plt.xlabel("Advertising Spend (in $)")
plt.ylabel("Revenue (in $)")
plt.grid(True)

# Adjust layout
plt.tight_layout()

# Display both plots
plt.show()
```
