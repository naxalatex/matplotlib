```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
crash_df = sns.load_dataset('car_crashes')
crash_df
```

```python
sns.histplot(crash_df['not_distracted'], bins=20)
```

```python
sns.jointplot(x= 'speeding', y= 'alcohol',data= crash_df, kind= 'kde')
```

```python
sns.histplot(crash_df['alcohol'])
```

```python
tip_df= sns.load_dataset('tips')
print(tip_df.to_string)
```

```python
sns.pairplot(tip_df, hue ='sex', palette= 'Blues', diag_kind= 'hist')
```
