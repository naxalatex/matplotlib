```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
x = np.array([1,2,3,4,5])
figure,axes = plt.subplots(2,2)

axes[0,0].plot(x, x*2, color='Red')
axes[0,0].set_title('x*2')

axes[0,1].plot(x,x**2, color='Blue')
axes[0,1].set_title('x**2')

axes[1,0].plot(x,x**3, color='Green')
axes[1,0].set_title('x**3')

axes[1,1].plot(x,x**4, color="Purple")
axes[1,1].set_title('x**4')

plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv('orders.csv')
print(df.to_string())
```

```python
cq = df.groupby('Category')['Quantity'].sum()
print(cq)
```

```python
plt.plot(cq.index, cq.values)
plt.show()
```

```python
plt.pie(cq,autopct='1.1f%%', labels=cq.index)
plt.show()
```

```python
plt.bar(cq.index, cq.values)
plt.show()
```

```python

# Task: Create a bar chart showing total sales for each country.
# Hint:
#   1. Add a 'Sales' column: df['Sales'] = df['Quantity'] * df['Price']
#   2. Group by 'Country' and sum 'Sales'
#   3. Sort values in descending order
#   4. Use plt.bar() to plot
#   5. Display the sales value above each bar
df['Sales'] = df['Quantity'] * df['Price']
print(df.to_string())
```

```python
cq = df.groupby('Country')['Sales'].sum()
cq.sort_values(ascending=True, inplace=True)    
print(cq)
```

```python
plt.barh(cq.index, cq.values)
plt.show()
```
