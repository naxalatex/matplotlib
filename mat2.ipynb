```python
import matplotlib.pyplot as plt
import numpy as np

years = np.array([2000, 2005, 2010, 2015, 2020])
population = np.array([1.0, 1.1, 1.2, 1.3, 1.38])

# Create figure and axis with dark backgrounds
fig, ax = plt.subplots(facecolor='black')      # Outer background
ax.set_facecolor('#1a1a1a')                    # Inner plot area (dark gray)

# Plot the data
ax.plot(years, population, marker='o', color='cyan', linewidth=2)

# Customize text
plt.title("Population growth of India (2000-2020)",
          family='Times New Roman', color='purple', fontsize=14, fontweight='bold')
plt.xlabel('Years', fontsize=15, family='Times New Roman',
           fontweight='bold', color='#4287f5')
plt.ylabel('Population', fontsize=10, family='Times New Roman',
           fontweight='bold', color='#42f548')

# Grid color adjustment
plt.grid(color='gray', linestyle='--', linewidth=0.5)

# Change tick color (so they’re visible on dark background)
ax.tick_params(colors='white')
plt.grid()
plt.xticks(years)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
x = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
productA = [120, 150, 170, 190, 210, 230]
productB = [100, 130, 160, 180, 200, 250]

# Plot both lines
plt.plot(x, productA, color='blue', linestyle='-', marker='o', linewidth=2, label='Product A')
plt.plot(x, productB, color='orange', linestyle='--', marker='s', linewidth=2, label='Product B')

# Title and labels
plt.title("Sales of Product A and B over 6 Months", fontsize=14, fontweight='bold', color='#06013b')
plt.xlabel('Months', fontsize=12, color='green', fontweight='bold')
plt.ylabel('Sales', fontsize=12, color='purple', fontweight='bold')

# Add grid and legend
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

courses = ['AI', 'ML', 'Data Science', 'Web Dev']
students = [50, 40, 60, 30]

# Create figure and axis with dark backgrounds
fig, ax = plt.subplots(facecolor='black')   # Outer background
ax.set_facecolor('#1a1a1a')                 # Inner plot area (dark gray)

# Bar chart
ax.bar(courses, students, color='#06013b', edgecolor='white', linewidth=1.5)

# Titles and labels (with bright text colors)
ax.set_title("Enrolled Students in Each Course",
             fontname='Monospace', fontsize=16, fontweight='bold', color='white')
ax.set_xlabel('Courses', color='cyan', fontname='Times New Roman', fontweight='bold', fontsize=15)
ax.set_ylabel('Students', color='lime', fontname='Times New Roman', fontweight='bold', fontsize=15)

# Make tick labels visible on dark background
ax.tick_params(colors='white', labelsize=12)

# Grid lines (optional)
ax.grid(axis='y', linestyle='--', alpha=0.5, color='gray')

plt.show()
```
