```python
import matplotlib.pyplot as plt
import numpy as np

years = np.array([2000, 2005, 2010, 2015, 2020])
population = np.array([1.0, 1.1, 1.2, 1.3, 1.38])

# Create figure and axis with dark backgrounds
fig, ax = plt.subplots(facecolor='black')      # Outer background
ax.set_facecolor('#1a1a1a')                    # Inner plot area (dark gray)

# Plot the data
ax.plot(years, population, marker='o', color='cyan', linewidth=2)

# Customize text
plt.title("Population growth of India (2000-2020)",
          family='Times New Roman', color='purple', fontsize=14, fontweight='bold')
plt.xlabel('Years', fontsize=15, family='Times New Roman',
           fontweight='bold', color='#4287f5')
plt.ylabel('Population', fontsize=10, family='Times New Roman',
           fontweight='bold', color='#42f548')

# Grid color adjustment
plt.grid(color='gray', linestyle='--', linewidth=0.5)

# Change tick color (so they’re visible on dark background)
ax.tick_params(colors='white')
plt.grid()
plt.xticks(years)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
x = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
productA = [120, 150, 170, 190, 210, 230]
productB = [100, 130, 160, 180, 200, 250]

# Plot both lines
plt.plot(x, productA, color='blue', linestyle='-', marker='o', linewidth=2, label='Product A')
plt.plot(x, productB, color='orange', linestyle='--', marker='s', linewidth=2, label='Product B')

# Title and labels
plt.title("Sales of Product A and B over 6 Months", fontsize=14, fontweight='bold', color='#06013b')
plt.xlabel('Months', fontsize=12, color='green', fontweight='bold')
plt.ylabel('Sales', fontsize=12, color='purple', fontweight='bold')

# Add grid and legend
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

courses = ['AI', 'ML', 'Data Science', 'Web Dev']
students = [50, 40, 60, 30]

# Create figure and axis with dark backgrounds
fig, ax = plt.subplots(facecolor='black')   # Outer background
ax.set_facecolor('#1a1a1a')                 # Inner plot area (dark gray)

# Bar chart
ax.bar(courses, students, color='#06013b', edgecolor='white', linewidth=1)

# Titles and labels (with bright text colors)
ax.set_title("Enrolled Students in Each Course",
             fontname='Monospace', fontsize=16, fontweight='bold', color='white')
ax.set_xlabel('Courses', color='cyan', fontname='Times New Roman', fontweight='bold', fontsize=15)
ax.set_ylabel('Students', color='lime', fontname='Times New Roman', fontweight='bold', fontsize=15)

# Make tick labels visible on dark background
ax.tick_params(colors='white', labelsize=10)

plt.grid(color='black', linestyle='--', linewidth=0.5)

# Grid lines (optional)
ax.grid(axis='y', linestyle='--', alpha=0.5, color='gray')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
brands = ['Samsung', 'Apple', 'Xiaomi', 'OnePlus', 'Others']
share = [30, 25, 20, 15, 10]

# Explode the slice with the highest share
explode = [0.1 if s == max(share) else 0 for s in share]

# Create figure with dark background
fig, ax = plt.subplots(figsize=(7,7), facecolor='black')
ax.set_facecolor('#1a1a1a')

# Plot pie chart
wedges, texts, autotexts = ax.pie(
    share,
    labels=brands,
    autopct='%1.1f%%',
    explode=explode,
    shadow=True,
    startangle=140,
    textprops={'color':'white'}  # make labels readable
)

# Title
ax.set_title("Mobile Brands Market Share", color='white', fontsize=14)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
categories = ['Rent', 'Food', 'Transport', 'Entertainment', 'Savings']
amount = [40000, 15000, 5000, 3000, 7000]

# Plot pie chart
plt.figure(figsize=(7,7))
plt.pie(amount, labels=categories, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title("Household Budget Distribution")
plt.axis('equal')  # Equal aspect ratio ensures pie is a circle
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
hours = [2, 3, 4, 5, 6, 7, 8]
scores = [50, 55, 60, 70, 75, 85, 90]

# Plot scatter
plt.figure(figsize=(8,5))
plt.scatter(hours, scores, color='purple', s=100)  # s is marker size
plt.title("Relationship Between Study Hours and Exam Scores", fontsize=14)
plt.xlabel("Study Hours", fontsize=12)
plt.ylabel("Exam Scores", fontsize=12)
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
height_A = [150, 160, 165, 170, 175]
weight_A = [50, 55, 60, 65, 70]

height_B = [152, 158, 168, 172, 178]
weight_B = [48, 53, 62, 66, 72]

# Plot scatter plots
plt.figure(figsize=(8,5))
plt.scatter(height_A, weight_A, color='blue', marker='o', s=100, label='Class A')
plt.scatter(height_B, weight_B, color='red', marker='s', s=100, label='Class B')
plt.grid(color='black', linestyle='--', linewidth=0.5)


# Labels and title
plt.xlabel("Height (cm)", fontsize=12)
plt.ylabel("Weight (kg)", fontsize=12)
plt.title("Height vs Weight of Class A and Class B", fontsize=14)
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
scores = [45, 50, 55, 60, 65, 70, 72, 75, 78, 80, 85, 90, 92, 95, 98]

# Create figure with black background
fig, ax = plt.subplots(figsize=(8,5), facecolor='black')
ax.set_facecolor('black')

# Plot histogram
ax.hist(scores, bins=5, color='skyblue', edgecolor='blue')  # edge in white for visibility

# Titles and labels in white
ax.set_title("Distribution of Test Scores", color='white', fontsize=14)
ax.set_xlabel("Scores", color='white', fontsize=12)
ax.set_ylabel("Number of Students", color='white', fontsize=12)

# Grid lines
ax.grid(axis='y', linestyle='--', alpha=0.7, color='gray')

plt.show()
```
